```python
shampoo_base=pd.concat([shampoo,shampoo.shift(4)],axis=1)
print(shampoo_base)
```

```python
shampoo_base.columns=['ActualSales','ForecastSales']
shampoo_base.dropna(inplace=True)
print(shampoo_base)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse_shampooerror=mean_squared_error(shampoo_base.ActualSales,shampoo_base.ForecastSales)
print(mse_shampooerror)
rmse=np.sqrt(mse_shampooerror)
print(rmse)
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(shampoo)
plt.show()
plot_pacf(shampoo)
plt.show()
```
